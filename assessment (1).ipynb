Question 1:
1. Create a Pandas Series s1 from the list [&#39;apple&#39;, &#39;banana&#39;, &#39;cherry&#39;, &#39;date&#39;, &#39;elderberry&#39;].
2. Assign custom indices [&#39;a1&#39;, &#39;b1&#39;, &#39;c1&#39;, &#39;d1&#39;, &#39;e1&#39;] to s1.
3. Update the value at index &#39;e1&#39; to &#39;elderflower&#39; and the value at index &#39;a1&#39; to &#39;apricot&#39;.
4. Extract and display all elements of s1 that have an index starting with &#39;c&#39; or &#39;d&#39;.

```python
import pandas as pd
```

```python
s1 = pd.Series(['apple', 'banana', 'cherry', 'date', 'elderberry'], index=['a1', 'b1', 'c1', 'd1', 'e1'])
```

```python
s1['e1'] = 'elderflower'
s1['a1'] = 'apricot'
filtered_s1 = s1.loc[['c1', 'd1']]
s1
filtered_s1
```

```python
data = {
    'Product_Name': ['Laptop', 'Smartphone', 'Shoes', 'Watch', 'TV'],
    'Category': ['Electronics', 'Electronics', 'Fashion', 'Fashion', 'Electronics'],
    'Units_Sold': [150, 300, 500, 200, 100],
    'Revenue': [120000, 180000, 75000, 50000, 100000],
    'Profit_Margin': [0.25, 0.15, 0.40, 0.30, 0.20]
}
products = pd.DataFrame(data)
```

```python
products['Total_Profit'] = products['Revenue'] * products['Profit_Margin']
products = products.drop(columns=['Profit_Margin'])
electronics_df = products[(products['Category'] == 'Electronics') & (products['Revenue'] > 100000)]
electronics_df
products
```

Question 5:
1. Use the Pokémon dataset (pokemon.csv) and perform the following operations:
o Display the names and types of Pokémon where Speed &gt; 60 and Attack &gt; 70.
o Filter Pokémon that are not Legendary and have a Type 1 of &#39;Water&#39;.
Display the Name, Type 1, and Legendary columns.
o Filter Pokémon with Defense &gt; 80 and Type 2 as &#39;Flying&#39;. Display their Name,
Type 1, Type 2, and Defense.

```python
pokemon = pd.read_csv('pokemon.csv')
filtered_pokemon = pokemon[(pokemon['Speed'] > 60) & (pokemon['Attack'] > 70)][['Name', 'Type 1', 'Type 2']]
non_legendary_water = pokemon[(pokemon['Legendary'] == False) & (pokemon['Type 1'] == 'Water')][['Name', 'Type 1', 'Legendary']]
defense_flying = pokemon[(pokemon['Defense'] > 80) & (pokemon['Type 2'] == 'Flying')][['Name', 'Type 1', 'Type 2', 'Defense']]
```
